## Line Charts

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('../Matplotlib_Bekzod_Ruzmetov/barcharts-us-music-sales-by-genre.csv', index_col='Year')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,4))

rock = ax.bar(df.index, 'Rock', data=df, label='Rock %')
jazz = ax.bar(df.index, 'Jazz', data=df, label='Jazz %')

ax.legend()
fig
```

```python
ax.set_title('Rock vs. Jazz', size=22)

ax.set_xticks(df.index)

spines = ['bottom', 'top', 'left', 'right']

for s in spines:
    ax.spines[s].set_visible(False)

ax.legend(frameon=False)

ax.tick_params(left='Off', bottom='Off')

fig
```

```python
for p in rock.patches:
    t = ax.annotate(str(p.get_height())+'%', (p))
```
