```python
import pandas as pd
```

```python
df = pd.read_csv('./Iris.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot('SepalLengthCm', data=df, label='Sepal')
ax.plot('PetalLengthCm', data=df, label='Petal')
ax.legend()
fig
```

```python
ax.set_title('Iris', size=22)
ax.set_xlabel('Index', size=14)
ax.set_ylabel('Length', size=14)

ax.tick_params(axis='both', labelsize=12, labelcolor='green')

from matplotlib.ticker import StrMethodFormatter

mft = StrMethodFormatter('{x:2f} inch')

ax.yaxis.set_major_formatter(mft)

fig
```

```python
line1 = ax.lines[0]

line1.set_marker('o')
line1.set_markevery(5)
line1.set_markerfacecolor('white')

line1.set(marker='o', markevery=5, markerfacecolor='white')
fig
```

```python
line1 = ax.lines[0]

line1.set(marker='o', markevery=5, markerfacecolor='white')
fig
```

```python
legend = ax.legend_

f1 = legend.get_lines()[0]
f1.set_marker('o')

legend.set_bbox_to_anchor([0.2,0.9])

legend.legendPatch.set_boxstyle('round, pad=0.5, rounding_size=1')

fig
```

```python
val = df['PetalLengthCm'].values

p_min = val.min()
p_max = val.max()

x_min = val.argmin()
x_max = val.argmax()
```

```python
ax.axvline(x_min, color='red', ls='dashed')
ax.axvline(x_max, color='green', ls='dashed')

fig
```

```python
min_annot = ax.annotate(str(p_min), xy=(x_min, p_min), xytext=(x_min+10, p_min+2), arrowprops=dict(arrowstyle='->'))

min_annot.set(size=22, color='white')
min_annot.set_bbox(dict(facecolor='red', boxstyle='round, pad=0.4', lw=0))

min_annot.arrow_patch.set_connectionstyle('angle, angleA=-90')
min_annot.arrow_patch.set(lw=2, linestyle='dashed', color='red')

max_annot = ax.annotate(str(p_max), xy=(x_max-10,p_max-2), arrowprops=dict(arrowstyle='->'))
max_annot.set(color='white', size=22)
max_annot.set_bbox(dict(facecolor='green', lw=0, boxstyle='round, pad=0.3'))

max_annot.arrow_patch.set_connectionstyle('angle, angleA=90')
max_annot.arrow_patch.set(color='green', linestyle='dashed', linewidth=2)



fig
```
