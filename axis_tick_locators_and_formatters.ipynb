## Axis Tick Locators and Formatters

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 fig, ax = plt.subplots()
fig
```

```python
ax.axis([0,11,0,11])
fig
```

```python
ax.set_xlim(0,12)
ax.set_ylim(0,8)
fig
```

```python
x = ax.xaxis
from matplotlib.ticker import AutoLocator, MaxNLocator, LinearLocator, MultipleLocator, FixedLocator, NullLocator
```

```python
auto_loc = AutoLocator()

x.set_major_locator(auto_loc)

fig
```

```python
max_t = MaxNLocator(5)
x.set_major_locator(max_t)
fig
```

```python
lin_loc = LinearLocator(5)
x.set_major_locator(lin_loc)
fig
```

```python
mul_loc = MultipleLocator(3)
x.set_major_locator(mul_loc)
fig
```

```python
fix_loc = FixedLocator(locs=[0,1,4,12])
x.set_major_locator(fix_loc)
fig
```

```python
null_loc = NullLocator()
x.set_major_locator(null_loc)
fig
```

```python
x.set_major_locator(auto_loc)
fig
```

```python
ax.set_xticks([0,1,4,12])
fig
```

```python
ax.set_xticks([0,1,4,12])
fig
```

```python
from matplotlib.ticker import NullFormatter, FixedFormatter, StrMethodFormatter, ScalarFormatter, PercentFormatter, IndexLocator
```

```python
null_form = NullFormatter()
x.set_major_formatter(null_form)
fig
```

```python
str_form = StrMethodFormatter('${x:.1f}')
x.set_major_formatter(str_form)
fig
```

```python
scal_form = ScalarFormatter()
x.set_major_formatter(scal_form)
fig
```

```python
null_form = NullFormatter()
x.set_major_formatter(null_form)
fig
```

```python
str_form = StrMethodFormatter('${x:.2f}')
x.set_major_formatter(str_form)
fig
```

```python
scal_form = ScalarFormatter()
x.set_major_formatter(scal_form)
fig
```

```python
per_form = PercentFormatter(xmax=12)
x.set_major_formatter(per_form)
fig
```

```python
ind_form = IndexError(list('abcdefghijklmnop'))
x.set_major_formatter(ind_form)
fig
```
